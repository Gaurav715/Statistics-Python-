# VISUALIZATION

1D

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
```

```python
d1 = np.loadtxt("example_1.txt")
d2 = np.loadtxt("example_2.txt")
print(d1.shape,d2.shape)
```

# Histrogram

```python
plt.hist(d1,label = 'D1')
plt.hist(d2,label = 'D2')
plt.legend()
plt.ylabel('counts')
```

```python
bins = np.linspace(min(d1.min(),d2.min()),max(d1.max(),d2.max()),50)
counts1m,_, _  =plt.hist(d1,label = 'D2',bins =10)
plt.hist(d2,label = 'D1',bins= bins)
plt.legend()
plt.ylabel('counts');
```

```python
bins = np.linspace(min(d1.min(),d2.min()),max(d1.max(),d2.max()),50)
counts1m,_, _  =plt.hist(d1,label = 'D2',bins =10,density = True)
plt.hist(d2,label = 'D1',bins= bins,density = True)
plt.legend()
plt.ylabel('Probablity');
```

```python
bins = np.linspace(min(d1.min(),d2.min()),max(d1.max(),d2.max()),50)
plt.hist([d1,d2],bins = bins , label = "Stacked",density= True,alpha =0.5)
plt.hist(d1,label = 'D2',bins =10,density = True,histtype= "step",lw= 1)
plt.hist(d2,label = 'D1',bins= bins,density = True,histtype = "step" ,ls = ":")
plt.legend()
plt.ylabel('Probablity');
```

```python
bins = np.linspace(min(d1.min(),d2.min()),max(d1.max(),d2.max()),50)
plt.hist([d1,d2],bins = bins , label = "Stacked",density= True,histtype ='barstacked',alpha =0.5)
plt.hist(d1,label = 'D2',bins =10,density = True,histtype= "step",lw= 1)
plt.hist(d2,label = 'D1',bins= bins,density = True,histtype = "step" ,ls = ":")
plt.legend()
plt.ylabel('Probablity');
```

# BEA SWARM PLOTS

```python
dataset =pd.DataFrame({
    "value": np.concatenate((d1,d2)),
    "type": np.concatenate((np.ones(d1.shape),np.zeros(d2.shape)))
})
dataset.info()
```

```python
sb.swarmplot(dataset["value"]);
```

```python
sb.swarmplot(x="type" ,y= "value",data = dataset,size =2);
```

# BOX_PLOT

```python
sb.boxplot(x= "type",y= "value",data= dataset,whis = 2.0)
sb.swarmplot(x= "type",y= "value",data= dataset,size = 2,color ="k",alpha = 0.3)
```

# VIOLIN_PLOT

```python
sb.violinplot(x= "type",y= "value",data= dataset,whis = 2.0)
sb.swarmplot(x= "type",y= "value",data= dataset,size = 2,color ="k",alpha = 0.3)
```

```python
sb.violinplot(x= "type",y= "value",data= dataset,inner ='quartile',bw =0.2 )
```

# Empirical Cumaltive Distribution Function

```python
sd1 = np.sort(d1)
sd2 = np.sort(d2)
cdf = np.linspace(1/d1.size, 1, d1.size)

plt.plot(sd1, cdf, label="D1 CDF")
plt.plot(sd2, cdf, label="D2 CDF")
plt.hist(d1, histtype="step", density=True, alpha=0.3)
plt.hist(d2, histtype="step", density=True, alpha=0.3)
plt.legend();
```
